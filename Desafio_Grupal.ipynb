## Cargar librerías y datos

Importar librerías

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%reload_ext autoreload
%autoreload 2

%matplotlib inline

pd.set_option("display.max_columns", 50)
```

Cargar dataset

```python
# Ubicación del dataset
location_df_train='Datasets/train.csv'
location_df_test='Datasets/test_x.csv'

# Llamar los datos en las variables
df_train=pd.read_csv(location_df_train)
df_test= pd.read_csv(location_df_test)
```

Mostrar 5 primeros registros

```python
df_train.head()
```

```python
df_test.head()
```

En ambos datasets se dropea el 'id'

```python
df_train.drop(columns='Id',inplace=True)
df_test.drop(columns='Id',inplace=True)
```

Se renombran las columnas para un mejor entendimiento de los datos

```python
# Nombres de las columnas para los dataset de train y test
column_names_train={'A':'EDAD',
                  'B':'TRABAJO',
                  'C':'ESTADO_CIVIL',
                  'D':'EDUCACION',
                  'E':'CREDITO_POR_DEFECTO',
                  'F':'PRESTAMO_VIVIENDA',
                  'G':'PRESTAMO_PERSONAL',
                  'H':'CONTACTO',
                  'I':'MES',
                  'J':'DIA',
                  'K':'DURACION',
                  'L':'CANT_CONTACT_REALIZADOS',
                  'M':'CANT_DIAS_ULT_CONTACTO',
                  'N':'CANT_CONTACT_PREV_REALIZADOS',
                  'O':'RESULT_CAMP_ANTERIOR',
                  'P':'TASA_VARIA_EMPLEO',
                  'Q':'INDICE_PREC_CONSUMIDOR',
                  'R':'INDICE_CONF_CONSUMIDOR',
                  'S':'TASA_INT_3_MESES',
                  'T':'CANT_EMPLEADOS',
                  'U':'TARGET'}
column_names_test=dict(list(column_names_train.items())[:-1])       

# Cambiar los nombres de las columnas de los datasets de train y test
df_train.rename(columns=column_names_train,inplace=True)
df_test.rename(columns=column_names_test,inplace=True)

# Mostrar el dataset de train (5 primeros registros) para verificar el cambio
df_train.head()
```
